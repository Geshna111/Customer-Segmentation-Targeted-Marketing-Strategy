```python
df = df.dropna(subset=['CustomerID'])
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

Data cleaning involved removing missing customer IDs, filtering out cancelled transactions, and eliminating invalid entries such as negative quantities. This ensures that the dataset reflects actual customer purchase behavior.
